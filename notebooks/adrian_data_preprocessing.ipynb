### Outliers

```python
"""Porcentaje de outliers de tempo"""
limit_6 = 200

tempo = data_pre["tempo"]
count_tempo = tempo[tempo > limit_6].count()

count_tempo/10000*100

#Resultado de 0.79% --> Standard Scaler
#Menor al 5% de outliers no hay ningun problema con scalers
```

```python
"""Porcentaje de outliers de loudness"""
limit_7 = -25

loudness = data_pre["loudness"]
count_loudness = loudness[loudness < limit_7].count()

count_loudness/10000*100

#Resultado de 2.48% --> Min-Max Scaler
#Menor al 5% de outliers no hay ningun problema con scalers
```

### Standard Scaler

```python
"""StandarScaler para tempo porque la data esta distribuida normal"""
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() # Instanciate StandarScaler

#Tempo
scaler.fit(data_pre[['tempo']]) # Fit scaler to data
data_pre['tempo'] = scaler.transform(data_pre[['tempo']]) # Use scaler to transform data
```

### MinMaxScaler

```python
"""MinMaxScaler para la skwed data"""
from sklearn.preprocessing import MinMaxScaler
mm_scaler = MinMaxScaler() # Instanciate MinMaxScaler

#Loudness
mm_scaler.fit(data_pre[['loudness']]) # Fit scaler to data
data_pre['loudness'] = mm_scaler.transform(data_pre[['loudness']]) # Use scaler to transform data

#Popularity
mm_scaler.fit(data_pre[['popularity']]) # Fit scaler to data
data_pre['popularity'] = mm_scaler.transform(data_pre[['popularity']]) # Use scaler to transform data
```
